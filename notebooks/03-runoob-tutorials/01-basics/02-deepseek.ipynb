# LangChain 集成 DeepSeek

本章节介绍如何在 LangChain 中集成和使用 DeepSeek 聊天模型。

通过 **langchain-deepseek** 扩展包，开发者可以快速接入 DeepSeek 提供的大语言模型服务。

DeepSeek 模型既支持官方托管 API，也支持通过 Ollama、Fireworks、Together 等平台进行本地或第三方推理部署。

## DeepSeek 介绍

DeepSeek 是一个开源大语言模型系列，支持聊天、推理、代码生成等能力。

在 LangChain 中，DeepSeek 主要通过 **ChatDeepSeek** 类进行调用。

### 集成信息

| 项目 | 说明 |
| --- | --- |
| 类名 | ChatDeepSeek |
| 安装包 | langchain-deepseek |
| 状态 | Beta（测试版） |
| 支持 JavaScript | 是 |
| 支持 Python | 是 |

### 模型能力支持

| 功能 | 是否支持 |
| --- | --- |
| 工具调用（Tool Calling） | ✅ 支持 |
| 结构化输出（Structured Output） | ✅ 支持 |
| 图片输入 | ❌ 不支持 |
| 音频输入 | ❌ 不支持 |
| 视频输入 | ❌ 不支持 |
| Token 流式输出 | ✅ 支持 |
| 原生异步调用 | ✅ 支持 |
| Token 使用统计 | ✅ 支持 |
| Logprobs | ❌ 不支持 |

DeepSeek API 使用与 OpenAI/Anthropic 兼容的 API 格式，通过修改配置，您可以使用 OpenAI/Anthropic SDK 来访问 DeepSeek API，或使用与 OpenAI/Anthropic API 兼容的软件。

| 参数 | 值 |
| --- | --- |
| base_url (OpenAI) | `https://api.deepseek.com` |
| base_url (Anthropic) | `https://api.deepseek.com/anthropic` |
| api_key | 点击链接申请 [API key](https://platform.deepseek.com/api_keys) |
| model <sup>*</sup> | `deepseek-v4-flash`、`deepseek-v4-pro`、`deepseek-chat` (将于 2026/07/24 弃用)、`deepseek-reasoner` (将于 2026/07/24 弃用) |

**术语：**

- **ChatDeepSeek**：LangChain 的 DeepSeek 聊天模型类
- **Tool Calling**：让 LLM 能够调用外部工具/函数的能力
- **结构化输出**：让 LLM 按指定格式（如 JSON）输出结果

## 创建 ChatDeepSeek 模型

安装完成后，可以通过 **ChatDeepSeek** 初始化模型：

````python
from langchain_deepseek import ChatDeepSeek

if api_key:
    # 创建模型
    llm = ChatDeepSeek(
        api_key=api_key,
        model="deepseek-v4-flash",
        temperature=0,
        max_tokens=None,
        timeout=None,
        max_retries=2
    )
    
    # 调用模型
    response = llm.invoke("你好，请介绍 LangChain")
    print(response.content)
else:
    print("需要配置 DEEPSEEK_API_KEY")
````

参数说明:

| 参数 | 说明 |
| --- | --- |
| api_key | 设置你申请的 API key，例如 "sk-xxx" |
| model | 指定模型名称，例如 deepseek-v4-flash |
| temperature | 控制随机性，越低结果越稳定 |
| max_tokens | 限制生成最大 Token 数量 |
| timeout | 请求超时时间 |
| max_retries | 失败后的最大重试次数 |

也可以使用 [init_chat_model()](https://www.runoob.com/langchain/langchain-init_chat_mode.html) 函数来调用：

````python
from langchain.chat_models import init_chat_model

# 指定了 model，返回固定模型
model = init_chat_model("deepseek:deepseek-v4-flash", temperature=0.7)
response = model.invoke("介绍langchain(一句话)")
print(response.content)
````

**术语：**

- **init_chat_model()**：LangChain 的统一模型初始化函数，自动根据模型标识选择对应的类
- **temperature**：控制输出随机性的参数（0=确定性，1=创造性）

## 调用 DeepSeek 模型

下面示例演示如何向 DeepSeek 模型发送聊天消息：

> 注意我这里测试，把 API Key 写在了测试文件中：
> ```
> llm = ChatDeepSeek(
>     api_key="sk-xxx",  # 设置你的 DeepSeek API Key
>     model="deepseek-v4-flash"
> )
> ```
> 
> 实际的生产环境请设置在.env 文件中。

````python
from langchain_deepseek import ChatDeepSeek

if api_key:
    # 注意：生产环境请使用 .env 文件管理 API Key
    llm = ChatDeepSeek(
        api_key=api_key,  # 设置你的 DeepSeek API Key
        model="deepseek-v4-flash",
        temperature=0,
        max_tokens=None,
        timeout=None,
        max_retries=2
    )

    messages = [
        (
            "system",
            "You are a helpful assistant that translates English to Chinese."
        ),
        (
            "human",
            "I love programming."
        ),
    ]

    ai_msg = llm.invoke(messages)
    print(ai_msg.content)
else:
    print("需要配置 DEEPSEEK_API_KEY")
````

运行后，模型会返回英文翻译后的中文结果。

```
我喜欢编程。
```

## LangChain + DeepSeek 完整测试实例

下面演示一个完整可运行的 LangChain + DeepSeek 示例。

### 创建测试文件

创建 test.py 文件：

````python
from langchain_deepseek import ChatDeepSeek
from langchain_core.messages import HumanMessage, SystemMessage

# =========================
# 创建 DeepSeek 模型
# =========================

llm = ChatDeepSeek(
    api_key=api_key,
    model="deepseek-v4-flash",
    temperature=0.7,
    max_tokens=1024,
    timeout=60,
    max_retries=3
)

# =========================
# 构造聊天消息
# =========================

messages = [
    SystemMessage(
        content="你是一名专业 Python 教师。"
    ),
    HumanMessage(
        content="请解释什么是 LangChain，并给出简单示例。"
    )
]

# =========================
# 调用模型
# =========================

response = llm.invoke(messages)

# =========================
# 输出结果
# =========================

print("AI 回复：")
print(response.content)
````

### 使用 deepseek-v4-pro 推理模型

LangChain 支持 DeepSeek 流式输出，可以使用 deepseek-v4-pro 推理模型：

````python
from langchain_deepseek import ChatDeepSeek
from langchain_core.messages import HumanMessage, SystemMessage

# =========================
# 创建 DeepSeek 模型
# =========================

llm = ChatDeepSeek(
    api_key=api_key,
    model="deepseek-v4-pro",
    temperature=0.7,
    max_tokens=1024,
    timeout=60,
    max_retries=3
)

# 流式输出
for chunk in llm.stream("请介绍 Python"):
    print(chunk.content, end="", flush=True)
````

**术语：**

- **流式输出（Streaming）**：逐个 token 返回结果，而不是等全部生成完才返回
- **推理模型（Reasoning Model）**：专门用于复杂推理任务的模型，如 deepseek-v4-pro

### 使用 PromptTemplate

结合 PromptTemplate 动态生成 Prompt：

````python
from langchain_deepseek import ChatDeepSeek
from langchain_core.prompts import ChatPromptTemplate

llm = ChatDeepSeek(
    api_key=api_key,
    model="deepseek-v4-pro",
    temperature=0.7,
    max_tokens=1024,
    timeout=60,
    max_retries=3
)

prompt = ChatPromptTemplate.from_template(
    "请详细解释：{topic}"
)

chain = prompt | llm

response = chain.invoke({
    "topic": "Transformer"
})

print(response.content)
````

**术语：**

- **PromptTemplate**：提示词模板，支持变量占位符（如 {topic}）
- **链式调用（|）**：LCEL 语法，将 prompt 和 llm 串联成链
